### Popularity based recommender system

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_ratings=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings.rename(columns={'Book-Rating':'Number of Ratings'},inplace=True)
```

```python
num_ratings
```

```python
mean_ratings=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].round(1).reset_index()
mean_ratings.rename(columns={'Book-Rating':'Average Ratings'},inplace=True)
mean_ratings
```

```python
total_ratings=num_ratings.merge(mean_ratings)
```

```python
total_ratings
```

```python
total_ratings[total_ratings['Book-Title'].str.contains('Gita')==True]
```

```python
# I want the most popular top 100 books
popular_books=total_ratings[total_ratings['Number of Ratings']>=150].sort_values('Average Ratings',ascending=False).head(100)
```

```python
popular_books
```

```python
popular_books=popular_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Number of Ratings','Average Ratings','Image-URL-M']]
popular_books.reset_index(drop=True,inplace=True)
popular_books.index=popular_books.index+1
```

```python
popular_books
```

### Collabrative based recommender system

```python
ratings_with_name
```

```python
x=ratings_with_name.groupby("User-ID").count()['Book-Rating']>125
badhiya_janta=x[x].index
```

```python
x
```

```python
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(badhiya_janta)]
```

```python
filtered_ratings
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>50
badhiya_books=y[y].index
```

```python
y
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(badhiya_books)]
```

```python
final_ratings
```

```python
final_ratings['Book-Title'].unique()
```

```python
len(final_ratings['Book-Title'].unique())
```

```python
final_ratings.duplicated().sum()
```

```python
ratings_pivot=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
ratings_pivot.fillna(0,inplace=True)
```

```python
ratings_pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score_ratings=cosine_similarity(ratings_pivot)
```

```python
similarity_score_ratings
```

```python
def recommend(book_name):
    index=np.where(ratings_pivot.index==book_name)[0][0]
    temp=sorted(list(enumerate(similarity_score_ratings[index])),key=lambda x:x[1],reverse=True)[1:7]
    for i in temp:
        print(ratings_pivot.index[i[0]])
```

```python
recommend("Harry Potter and the Chamber of Secrets (Book 2)")
```

```python
def recommend2(book_name):
    index=np.where(ratings_pivot.index==book_name)[0][0]
    temp=sorted(list(enumerate(similarity_score_ratings[index])),key=lambda x:x[1],reverse=True)[1:9]
    data=[]
    for i in temp:
        temp2=filtered_ratings[filtered_ratings['Book-Title']==ratings_pivot.index[i[0]]].drop_duplicates(['Book-Title'])
        item=[]
        item.extend(list(temp2['Book-Title'].values))
        item.extend(list(temp2['Book-Author'].values))
        item.extend(list(temp2['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend2("Harry Potter and the Chamber of Secrets (Book 2)")
```
